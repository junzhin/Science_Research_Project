```python
import pandas as pd
import numpy as np
```

```python
acc_file_path = r'/Volumes/mac_data/2022coding_files/Summer_project/plots_results/Alexnet_office31_acc.csv'
test_loss_file_path = r'/Volumes/mac_data/2022coding_files/Summer_project/plots_results/Alexnet_office31_test_loss.csv'
acc_df = pd.read_csv(acc_file_path)
test_loss_file_path = pd.read_csv(test_loss_file_path)
```

```python
test_loss_file_path
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (6,5))
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(test_loss_file_path['Random']),color ='red',label = "Random" )
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(test_loss_file_path['Masked']), color ='black', label = "Masked")
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(test_loss_file_path['Original']),color ='blue', label = "Original")
plt.ylabel("Test Loss")
plt.xlabel('Epochs')
plt.legend()
plt.savefig("Test_Loss.png")
```

```python
fig = plt.figure(figsize = (6,5))
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(acc_df['Random']),color ='red',label = "Random" )
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(acc_df['Masked']), color ='black', label = "Masked")
plt.plot(np.array([i for i in range(len(test_loss_file_path))]),np.array(acc_df['Original']),color ='blue', label = "Original")
plt.ylabel("Accuarcy")
plt.xlabel('Epochs')
plt.legend()
plt.savefig("Accuarcy.png")
```
